# A Wordcloud demo

Simple Python script to convert a (long) text into a WordCloud image.

We will use the `wordcloud` package for this, as well as `matplotlib` for plotting.

```python
import matplotlib.pyplot as plt
import requests
from wordcloud import WordCloud
```

the `requests` package allow to retrieve a remote file and store its content into a variable. We will fetch the "Origin of Species" by Charles Darwin.

```python
# Fetch text from the URL
url = "https://raw.githubusercontent.com/telatin/learn_bash/master/files/origin.txt"
response = requests.get(url)

# Response. Code 200 means OK. 404 as you know would mean not found.
response
```

```python
# Get the content
text = response.text

# Show some characters from "text"
text[80:250]
```

With a single command you can now generate a **WordCloud** object, running the *generate* method on the "text" object.

```python
# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
```

finally, we can use MatPlotLib to generate the plot. We can configure the size, and hiding the axis is a good idea.

```python
# Display the Word Cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud)
plt.axis('off')  # Hide the axis
plt.show()
```
